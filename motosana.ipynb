```python
import pandas as pd

wekaweka = pd.read_csv("air_quality_no2.csv",index_col=0, parse_dates=True)
#The new column =>london_mg_per_cubic 
#Existing column => station_london
wekaweka["london_mg_per_cubic"] = wekaweka["station_london"] * 1.882

#checking ratios of values in Paris and Antwerp

#New column => ratio_paris_antwerp 

wekaweka["ratio_paris_antwerp"] = (
    wekaweka["station_paris"]/wekaweka["station_antwerp"]
)


wekaweka_rename = wekaweka.rename(

    columns={
        #old_name | New_name
        "station_antwerp":"Dan",
        "station_paris":"Dotty",
         "station_london":"Ronaldo"
    }

)

wekaweka_rename.head()
```
